# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

````python
# Insert your code here...
data_movies_copy = data_movies.copy()  # Creating a copy of dataset for future purpose
print(data_movies is data_movies_copy) # This is to ensure both datasets are at different memory locations
````

````python
data_movies.info()
````

#### Rename the required Columns


````python
data_movies.rename(columns={'Runtime (Minutes)' : 'RuntimeMinutes'}, inplace=True)
data_movies.rename(columns={'Revenue (Millions)' : 'RevenueMillions'}, inplace=True)
data_movies.info()
````

- We are renaming the columns to have single value so that for future analysis we can use . operator to access the columns insted of specifying them in [].

````python
data_movies.drop(['Description'], axis=1, inplace=True)
````

- **Description** column is dropped because we cannot use this coulumn to get much insights during our analysis

````python
data_movies.info()
````

- ```Working on Actors Column```
- The actors column has 4 actor details per each row so we are going to split them into 4 different columns for bettwe analysis.
- Actors column will be splited into Actor1, Actor2, Actor3 and Actor4. Post split Actor column will be deleted from the dataframe as it will be a redundent data


````python
data_movies.Actors
````

````python
# data_movies[['Actor1', 'Actor2', 'Actor3', 'Actor4']] = data_movies.Actors.str.split(',')
# data_movies.Actors.str.split(',')
actors = data_movies.Actors.str.split(',', expand=True)
actors
data_movies['Actor1'] = actors[0]
data_movies['Actor2'] = actors[1]
data_movies['Actor3'] = actors[2]
data_movies['Actor4'] = actors[3]
print(data_movies.info())
data_movies.head()
data_movies[data_movies.Actor4.isnull()]
# data_movies.iloc[622]
````

- The column **Actors** will be **dropped** because they **don't provide** any additional **insights** into the data


````python
data_movies.drop(['Actors'], axis=1, inplace=True)
````

````python
data_movies.info()
````

- ```Working on Genre Column```
- The genre column has 3 genre details per each row so we are going to split them into 3 different columns for bettwe analysis.
- Genre column will be splited into Genre1, Genre2 and Genre3. Post split Genre column will be deleted from the dataframe as it will be a redundent data


````python
genre = data_movies.Genre.str.split(',' , expand=True)
data_movies['Genre1'] = genre[0]
data_movies['Genre2'] = genre[1]
data_movies['Genre3'] = genre[2]
````

````python
data_movies.info()
````

- The column **Genre** will be **dropped** because they **don't provide** any additional **insights** into the data
- The columns **Genre2 and Genre3** will also be be dropped as we can can derive most of the insights from **Genre1** column

````python
data_movies.drop(['Genre'], axis=1, inplace=True)
data_movies.drop(['Genre2', 'Genre3'], axis=1, inplace=True)
````

- Renaming the **Genre1** column to **Genre** as now we have only single Genre column in our dataset

````python
data_movies.rename(columns={'Genre1' : 'Genre'}, inplace=True)
````

````python
data_movies.info()
````

- Rearranging the columns in dataframe

````python
data_movies = data_movies[['Rank', 'Title', 'Genre', 'Director', 'Year', 'RevenueMillions', 
                          'Rating', 'Votes', 'Metascore', 'RuntimeMinutes', 'Actor1', 'Actor2', 
                          'Actor3', 'Actor4']]
````

````python
data_movies.info()
````

- ```Working on Revenue Column```
- The Revenue column has missing data hence replacing this missing value with median of Revenue column


````python
# Replacing missing values (NaN) for revenue with median 
# data_movies[data_movies['Revenue (Millions)'].isnull()]
data_movies['RevenueMillions'].fillna(data_movies['RevenueMillions'].median(), inplace=True)
````

````python
# Replacing missing values (NaN) for Metascore with median 
# data_movies[data_movies.Metascore.isnull()]
data_movies.Metascore.fillna(data_movies.Metascore.median(), inplace=True)
````

````python
data_movies.info()
````

- **Rounding** the **values** of **RevenueMillions** columns to 2 decimal places.

````python
data_movies = data_movies.round(decimals=2)
data_movies.head()
````

### Working with Year and Revenue

- Top 100 movies interms of **Revenue**

````python
top_100_movies = data_movies.iloc[data_movies.RevenueMillions.sort_values(ascending=False)[:100].index, :]
top_100_movies
````
