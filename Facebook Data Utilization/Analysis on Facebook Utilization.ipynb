# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

```python
data_copy = data.copy()
```

###  Deleting the rows wich has no gender value as we cannot predict the gender from the available data

```python
data.drop(data[data.gender.isna()].index, inplace=True)
```

```python
data.info()
```

### Deleting the rows which has no tenure information

```python
data.drop(data[data.tenure.isna()].index, inplace=True)
```

```python
data.info()
```

```python
data.head()
```

### **Below is the condition used to delete inactive profiles from the data**
- Profile with tenure more than 3 months and has not initiated any friend requests and has no friends
- Profiles with tenure more than 3 months and has not liked any post or received any likes for their post

```python
data[(data.friend_count == 0) & (data.friendships_initiated == 0) & (data.tenure > 90)]
```

```python
data.drop(data[(data.friend_count == 0) & (data.friendships_initiated == 0) & (data.tenure > 90)].index,\
          inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
data[(data.likes == 0) & (data.likes_received == 0) & (data.mobile_likes ==0) & \
(data.mobile_likes_received ==0) & (data.www_likes == 0) & (data.www_likes_received == 0) & \
(data.tenure > 90)]
```

```python
data.drop(data[(data.likes == 0) & (data.likes_received == 0) & (data.mobile_likes ==0) & \
(data.mobile_likes_received ==0) & (data.www_likes == 0) & (data.www_likes_received == 0) & \
(data.tenure > 90)].index, inplace=True)
```

```python
data.info()
```

```python
data.head()
```

### Making userid as index for the dataframe as userid can be uniquely used to find each row 

```python
data.set_index('userid', inplace=True)
```

```python
data.info()
```

### Creating a new column age group for detailed analysis

```python
labels = ['10-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100', '101-110', '111-120']
data['age_group'] = pd.cut(data.age, bins=np.arange(10, 121, 10), labels=labels, right=True)
```

```python
data.sort_index().head()
```

### Dropping dob_day, dob_month and dob_year columns as we can derive most of our insights from age and age group columns

```python
data.drop(['dob_day', 'dob_month', 'dob_year'], axis=1, inplace=True)
```

```python
data.info()
```

```python
data[(data.age >= 10) & (data.age <= 20) & (data.friend_count == 0)]
```

```python
data.info()
```

```python
plt.rcParams['figure.figsize'] = 12, 7
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

### Distribution of users based on gender

```python
data.gender.value_counts().plot(kind='bar', figsize=(4,8), colormap='Dark2', fontsize=15)
plt.xlabel('Gender', fontsize=15)
plt.ylabel('Total no.of people', fontsize=15)
plt.title('Barchart showing number of people as per gender', fontsize=15);
```

```python
data['gender'].value_counts().plot(kind='pie', fontsize=14, autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                       shadow=True, startangle=160, figsize=(10,10), cmap='inferno', legend=True)

plt.ylabel('% Distribution', fontsize=16)
plt.title('Donut Plot showing the distribution of users based on gender', fontsize=15);
```

```python
data.gender.value_counts().plot(kind='pie',  fontsize=14, autopct='%5.1f%%', figsize=(8,8),
                               shadow=True, cmap='summer', explode=[0.05,0.05])
plt.ylabel('Gender Distribution', fontsize=15)
plt.title('Pie Plot showing the distribution of users based on gender', fontsize=15);
```

- From this visualization we can depict that male users are more in number than female users on facebook

### Distribution of uses based on age group

### Distribution of tottal likes, mobile likes and website likes based on gender

```python
fig, axes = plt.subplots(6,2)
fig.set_size_inches(30, 80)
fig.align_labels()

p1 = data.groupby(['gender'])['likes'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='Total no.of likes sent', ax=axes[0,0])

p2 = data.groupby(['gender'])['likes'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='Total no.of likes sent', ax=axes[0,1])

p3 = data.groupby(['gender'])['mobile_likes'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='No.of likes sent from mobile', ax=axes[1,0])

p4 = data.groupby(['gender'])['mobile_likes'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='No.of likes sent from mobile', ax=axes[1,1])

p5 = data.groupby(['gender'])['www_likes'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='No.of likes sent from website', ax=axes[2,0])

p6 = data.groupby(['gender'])['www_likes'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='No.of likes sent from website', ax=axes[2,1])

p7 = data.groupby(['gender'])['likes_received'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='Total no.of likes received', ax=axes[3,0])

p8 = data.groupby(['gender'])['likes_received'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='Total no.of likes recceived', ax=axes[3,1])

p9 = data.groupby(['gender'])['mobile_likes_received'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='No.of likes received from mobile', ax=axes[4,0])

p10 = data.groupby(['gender'])['mobile_likes_received'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='Total no.of likes recceived from mobile', ax=axes[4,1])

p11 = data.groupby(['gender'])['www_likes_received'].sum().sort_values(ascending=False).plot(
                 kind='bar',   fontsize=25, ylabel='No.of likes received from website', ax=axes[5,0])

p12 = data.groupby(['gender'])['www_likes_received'].sum().sort_values(ascending=False).plot(
                         kind='pie',  autopct='%5.1f%%', fontsize=25,explode=[0.05,0.05],
                                 cmap='summer', ylabel='Total no.of likes recceived from website', ax=axes[5,1])
```

**From the above visualiztion we can get the below insights**
- Female users has sent 50% more likes compared to male users which shows that female users are more active in responding to a post
- 71% of the overall likes sent from website are from female users
- Female users has received 72% of the overall likes from both mobile app and website
- Female users contribute to maximum share of likes received and sent when compared to the male users
